```python
import numpy as np
import numpy.matlib
import jax
import jax.numpy as jnp
from jax import jit, grad, random, jacfwd, value_and_grad
from jax.ops import index, index_add, index_update
from functools import partial
import time
from utilfuncs import MMA, ThermalMesher, computeLocalElements, \
        computeFilter,applySensitivityFilter
from mmaOptimize import optimize
from fiberCompliance import FiberCompliance
import matplotlib.pyplot as plt
rand_key = random.PRNGKey(0);
```

### Mesh

```python
nelx, nely = 20, 12;
elemSize = np.array([1., 1.])
mesh = {'nelx':nelx, 'nely':nely, 'elemSize':elemSize,\
        'ndof':(nelx+1)*(nely+1), 'numElems':nelx*nely};
```

### Material

```python
material = {'kx':1., 'ky':2., 'alpha':1.e-4, 'penal':3};
FC = FiberCompliance(mesh);
Ktemplt = FC.getThermalTemplates();
```

### Filter

```python
filterRadius = 1.5;
H, Hs = computeFilter(mesh, filterRadius);
ft = {'type':1, 'H':H, 'Hs':Hs};
```

### Compliance minimization

```python
# mid edge load
force = np.zeros((mesh['ndof'], 1))
force[np.arange(int(nelx * (nely + 1)), int((nelx + 1) * (nely + 1)))] = 1
fixed = np.arange(0, nely, 1) #int(nely / 2 + 1 - nely / 20);
free = np.setdiff1d(np.arange(mesh['ndof']),fixed);
bc = {'heat':force, 'fixedTempNodes':fixed, 'freeTempNodes':free};

```

### Constraint

```python
globalVolumeConstraint = {'isOn':True, 'vf':0.5};
```

### Optimize

```python
optimizationParams = {'maxIters':150,'minIters':50,'relTol':0.02};
```

```python
class ThermalComplianceMinimizer:
    def __init__(self, mesh, bc, material, \
                 globalvolCons):
        self.mesh = mesh;
        self.material = material;
        self.bc = bc;
        M = ThermalMesher();
        self.edofMat, self.idx = M.getMeshStructure(mesh);
        self.D0 = M.getD0();
        self.globalVolumeConstraint = globalvolCons;
        self.objectiveHandle = jit(value_and_grad(self.computeObjective))
        self.consHandle = self.computeConstraints;

        self.numConstraints = 1;
    #-----------------------#
    def computeObjective(self, desvar):
        rho = desvar[0:self.mesh['numElems']];
        theta = desvar[self.mesh['numElems']:];
        @jit
        def SIMPMaterial(rho):
            k = self.material['kx']*(rho+0.01)**self.material['penal'];
            return k;
        Y = SIMPMaterial(rho);
        #-----------------------#
        @jit
        def computeRotations(theta):
            Q00 = jnp.array([self.material['kx']*jnp.cos(theta)**2 + \
                             self.material['ky']*jnp.sin(theta)**2]).reshape(-1)
            Q01 = jnp.array([0.5*(self.material['ky']-self.material['kx'])*\
                             jnp.sin(2*theta)]).reshape(-1)
            Q11 = jnp.array([self.material['ky']*jnp.cos(theta)**2 + \
                             self.material['kx']*jnp.sin(theta)**2]).reshape(-1)
            return Q00, Q01, Q11
#         theta = 0.25*np.pi*jnp.ones((mesh['numElems']));
        Q00, Q01, Q11 = computeRotations(theta);
        #-----------------------#
        @jit
        def assembleK(k):
            K = jnp.zeros((self.mesh['ndof'],)*2);
            D0 =jnp.einsum('k,ij->ijk',Q00,Ktemplt['00']) + \
                jnp.einsum('k,ij->ijk',Q01,Ktemplt['01']) +\
                jnp.einsum('k,ij->ijk',Q11,Ktemplt['11'])
            sK = jnp.einsum('i,jki->jki',k,D0).T.flatten()
            K = jax.ops.index_add(K, self.idx, sK);
            return K;
        K = assembleK(Y);
        #-----------------------#
        @jit
        def solve(K): 
            
            Kfree = K[self.bc['freeTempNodes'],:]\
                    [:,self.bc['freeTempNodes']];
            ffree = self.bc['heat'][self.bc['freeTempNodes']] 
    
            u_free = jax.scipy.linalg.solve(Kfree, ffree);
    
            u = jnp.zeros((self.mesh['ndof']));
            u = jax.ops.index_add(u, self.bc['freeTempNodes'],\
                                  u_free.reshape(-1));
            return u;
        u = solve(K);
        #-----------------------#
        @jit
        def computeCompliance(K, u):
            J = jnp.dot(u.T, jnp.dot(K,u));
            return J;
        J = computeCompliance(K, u);
        return J; 

    #-----------------------#
    def computeConstraints(self, desvar, epoch): 
        rho = desvar[0:self.mesh['numElems']];
        theta = desvar[self.mesh['numElems']:];
        
        @jit
        def computeGlobalVolumeConstraint(rho):
            vc = jnp.mean(rho)/self.globalVolumeConstraint['vf'] - 1.;
            return vc;

        if(self.globalVolumeConstraint['isOn']): 
            c, gradc = value_and_grad(computeGlobalVolumeConstraint)\
                                    (rho);
            gradc = np.hstack(( gradc.reshape((1,-1)),\
                               np.zeros((1,self.mesh['numElems'])) )) # zero sens wrt theta

            c, gradc = c.reshape((1,1)), gradc;

        return c, gradc
    #-----------------------#
    def TO(self, optimizationParams, ft):
        rho = jnp.ones((self.mesh['numElems'])); 
        theta = 0.25*np.pi*jnp.ones((self.mesh['numElems'])); 
        desvar = jnp.hstack((rho,theta))
        loop = 0; 
        change = 1.;
        m = self.numConstraints; # num constraints
        n = 2*mesh['numElems'] ;
        mma = MMA();
        mma.setNumConstraints(self.numConstraints);
        mma.setNumDesignVariables(n);
    
        minBound = np.vstack((  np.zeros((self.mesh['numElems'],1)), \
                        0.*np.pi*np.ones((self.mesh['numElems'],1))  ));
        maxBound = np.vstack((  np.ones((self.mesh['numElems'],1)), \
                        1.*np.pi*np.ones((self.mesh['numElems'],1))  ));
    
        mma.setMinandMaxBoundsForDesignVariables(minBound,maxBound);

        xval = desvar[np.newaxis].T 
        xold1, xold2 = xval.copy(), xval.copy();
        mma.registerMMAIter(xval, xold1, xold2);
        mma.setLowerAndUpperAsymptotes(np.ones((n,1)), np.ones((n,1)));
        mma.setScalingParams(1.0, np.zeros((m,1)), \
                             10000*np.ones((m,1)), np.zeros((m,1)))
        mma.setMoveLimit(0.2);

        mmaTime = 0;

        t0 = time.perf_counter();

        while( (change > optimizationParams['relTol']) \
               and (loop < optimizationParams['maxIters'])\
               or (loop < optimizationParams['minIters'])):
            loop = loop + 1;

            J, dJ = self.objectiveHandle(desvar); 

            vc, dvc = self.consHandle(desvar, loop);
            dJ, dvc = np.array(dJ), np.array(dvc)

            dJrho, dvcrho = applySensitivityFilter(ft, rho, dJ[0:self.mesh['numElems']], \
                                                      dvc[0,0:self.mesh['numElems']])
            dJ[0:self.mesh['numElems']] = dJrho
            dvc[0,0:self.mesh['numElems']] = dvcrho
               
            J, dJ = J, dJ[np.newaxis].T

            tmr = time.perf_counter();
            mma.setObjectiveWithGradient(J, dJ);
            mma.setConstraintWithGradient(vc, dvc);

            xval = desvar.copy()[np.newaxis].T;


            mma.mmasub(xval);

            xmma, _, _ = mma.getOptimalValues();
            xold2 = xold1.copy();
            xold1 = xval.copy();
            desvar = xmma.copy().flatten()
            mma.registerMMAIter(desvar, xval.copy(), xold1.copy())
            rho = desvar[0:self.mesh['numElems']];
            theta = desvar[self.mesh['numElems']:];
        
            mmaTime += time.perf_counter() - tmr;

            status = 'Iter {:d}; J {:.2F}; vf {:.2F}'.\
                    format(loop, J, jnp.mean(rho));
            print(status)
            if(loop%10 == 0):
                plt.imshow(-np.flipud(rho.reshape((mesh['nelx'], \
                         mesh['nely'])).T), cmap='gray');
                plt.quiver(np.sin(theta.reshape(nelx, nely).T),\
                           np.cos(theta.reshape(nelx, nely).T), color = 'red')
                plt.title(status)
                plt.show()
                plt.figure()
                plt.hist(theta)
                plt.show()
        totTime = time.perf_counter() - t0;

        print('total time(s): ', totTime);  
        print('mma time(s): ', mmaTime);
        print('FE time(s): ', totTime - mmaTime);
```
